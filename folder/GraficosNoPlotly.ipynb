# Gr√°ficos Interativos Plotly

```python
import pandas as pd
import plotly.express as px
import numpy as np
#ver todas as colunas e todas as linhas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.options.display.float_format = '{:.2f}'.format
```

```python
df = pd.read_csv('dados.csv')
```

```python
df.head()
```

```python
df.tail()
```

## Gr√°fico de Barras

```python
df_agrupado_uf = df.groupby('UF_Cliente')['Saldo_Devedor'].mean()

df_agrupado_uf = df_agrupado_uf.sort_values(ascending=False)

df_agrupado_uf = df_agrupado_uf.reset_index()
```

```python
df_agrupado_uf
```

```python
px.bar(df_agrupado_uf, x='UF_Cliente', y='Saldo_Devedor')
```

```python
df_agrupado_UF_Genero = df.groupby(['UF_Cliente', 'Sexo'])['Saldo_Devedor'].mean().sort_values(ascending=False).reset_index()
```

```python
df_agrupado_UF_Genero
```

```python
px.bar(df_agrupado_UF_Genero, x='UF_Cliente', y='Saldo_Devedor', color='Sexo')
```

```python
px.bar(df_agrupado_UF_Genero, x='UF_Cliente', y='Saldo_Devedor', color='Sexo', barmode = 'group')
```

```python
#inverter as posi√ß√µes dos eixos, se n√£o sair√° em branco
px.bar(df_agrupado_UF_Genero, x='Saldo_Devedor', y='UF_Cliente', color='Sexo', barmode = 'group', orientation = 'h')
```
